```python
driver = webdriver.Chrome()
```

```python
url = "https://www.billboard.com/charts/greatest-hot-100-singles"
driver.get(url)

all_items_billboard=[]

items = driver.find_elements_by_class_name('chart-list-item')

for item in items:
    song={}
    song["songs"]= item.find_element_by_class_name('chart-list-item__title-text').text
    song["artists"] = item.find_element_by_class_name('chart-list-item__artist').text
    song["rankings"]= item.find_element_by_class_name('chart-list-item__rank').text
    all_items_billboard.append(song)
    
#     print(songs.text,artists.text, rankings.text)
    
all_items_billboard
```
